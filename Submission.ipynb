## **Hypothesis Testing**

####  Hypothesis statement -1 <br>
<ul>
<li><b> NUll Hypothesis (H)</b>: Feature has no predictive power.</li>
<li><b>Alternate Hypothesis (H)</b>:Feature significantly predicts credit score.</li>
</ul>

```python
df_test=df.copy()
df_test['Poor_score'] = (df['Credit_Score'] == 'Poor').astype(int)
```

```python
import statsmodels.api as sm
X = df_test[['Annual_Income', 'Outstanding_Debt', 'Num_Bank_Accounts', 'Num_Credit_Inquiries']]
X = sm.add_constant(X)
y = df_test['Poor_score']

model = sm.Logit(y, X).fit()
print(model.summary())
```

since p<0.005 therefore,All four features are statistically significant predictors of poor credit score.

####  Hypothesis statement -2 <br>
<ul>
<li><b> NUll Hypothesis (H)</b>: payment behavior is the same across credit score groups.</li>
<li><b>Alternate Hypothesis (H)</b>: payment behavior differs across groups.</li>
<li><b>Test</b>: Chisquare</li>
<li><b>aplha</b>= 0.05</li>
</ul>

```python
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(df['Credit_Score'], df['Payment_Behaviour'])

# Run Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Test - Payment_Behaviour vs Credit_Score")
print("Chi2 Statistic:", chi2, "p-value:", p)



alpha = 0.05
if p < alpha:
    print("Reject H0 --> Payment behavior differs across credit score groups")
else:
    print("Fail to reject H0 --> No significant difference in payment behavior across groups")

```

####  Hypothesis statement -3 <br>
<ul>
<li><b> NUll Hypothesis (H)</b>: No multicollinearity among predictors..</li>
<li><b>Alternate Hypothesis (H)</b>: Predictors grous shows multicollinearity .</li>
<li><b>Test</b>: VIF</li>
</ul>

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df[['Outstanding_Debt', 'Num_Bank_Accounts', 'Num_Credit_Inquiries']]
X = sm.add_constant(X)

vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

Since VIF is less than 5 for all features thererfore there is no strong multicolinearity among predictors.
