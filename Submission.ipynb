### Univariate Analysis

#### Chart - 1

```python
# chart code
cols= ['Age', 'Annual_Income', 'Outstanding_Debt', 
        'Credit_Utilization_Ratio', 'Monthly_Balance', 'Num_Bank_Accounts']
plt.figure(figsize=(15,10))
for i, col in enumerate(cols, 1):
    plt.subplot(2, 3, i) 
    sns.histplot(df[col], bins=30, color="#FFC7A7", edgecolor="black")
    plt.title(col)
plt.tight_layout()
plt.show()




```

#### Chart - 3

```python
cols=['Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour']
fig, axes = plt.subplots(2, 2, figsize=(15,10))  
for ax, col in zip(axes.flatten(), cols):
    sns.countplot(x=df[col], ax=ax)
    ax.set_title(f"Distribution of {col}")
    ax.set_xlabel("")
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=45) 
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

Answer :<br>
TO inspect whether any categorical features contain imbalance in dataset.since the imbalance dataset lead to affect model learning.
